## 1. Summary Statistics
### (a) Calculate and display the mean and volatility of each asset’s excess return. (Recall we use volatility to refer to standard deviation.)

```python
import pandas as pd
import numpy as np
```

```python
excess_return = pd.read_excel("../data/multi_asset_etf_data.xlsx", sheet_name="excess returns")
excess_return
```

```python
summary = excess_return.describe().iloc[[1, 2],1:]
summary.index = ["mean", "volatility"]
summary
```

```python
summary.loc["Sharpe Ratio"] = summary.loc["mean"]/summary.loc["volatility"]
summary
```

### (b) Which assets have the best and worst Sharpe ratios

```python
print("Sharpe Ratio Max: %s \nSharpe Ration Min: %s" % (summary.idxmax(axis=1)["Sharpe Ratio"],summary.idxmin(axis=1)["Sharpe Ratio"]))
```
